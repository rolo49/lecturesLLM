# Linear regression with gradient descent

__Simulate data__

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Set seed
np.random.seed(123)

# Sample size
n = 10000

# Independent variable
x = np.random.uniform(-10, 10, n)

# True intercept
alpha_true = 2

# True slope
beta_true = 3

# Error term
eps = np.random.normal(0, 1, n)

# Create dependent variable
y = alpha_true + beta_true * x + eps
```

__Gradient descent__

```python
# Initialise parameters
alpha = 0
beta = 0
learning_rate = 0.01
num_epochs = 1000

# Gradient descent loop
for i in range(1, num_epochs + 1):

    # Predict values
    y_pred = alpha + beta * x

    # Compute residual
    res = y - y_pred

    # Gradients
    grad_alpha = -2 * np.mean(res)
    grad_beta = -2 * np.mean(res * x)

    # Update parameters
    alpha = alpha - learning_rate * grad_alpha
    beta = beta - learning_rate * grad_beta

    # Status update
    if i % (num_epochs // 10) == 0:
        print(f"Epoch: {i}  Estimated alpha: {alpha:.6f}  Estimated beta: {beta:.6f}")

# Output the final estimated parameters
print(f"\nFinal estimated alpha: {alpha:.6f}")
print(f"Final estimated beta:  {beta:.6f}")
```

__Estimate the model with scikit-learn's `LinearRegression`__

```python
model = LinearRegression()
model.fit(x.reshape(-1, 1), y)

print(f"Intercept: {model.intercept_:.6f}")
print(f"Slope:     {model.coef_[0]:.6f}")
```
