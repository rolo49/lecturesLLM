# Logistic Regression with gradient descent

__Simulate data__

```python
import numpy as np
from sklearn.linear_model import LogisticRegression

# Set seed
np.random.seed(123)

# Sample size
n = 10000

# Independent variable
x = np.random.uniform(-10, 10, n)

# True intercept
alpha_true = -3

# True slope
beta_true = 0.8

# Compute probability with the logistic function
p = 1 / (1 + np.exp(-(alpha_true + beta_true * x)))

# Simulate binary outcomes
y = np.random.binomial(1, p, n)
```

__Gradient descent__

```python
# Initialise parameters
alpha = 0
beta = 0
learning_rate = 1
num_epochs = 1000

# Gradient descent loop
for i in range(1, num_epochs + 1):

    # Predict probabilities
    p = 1 / (1 + np.exp(-(alpha + beta * x)))

    # Compute gradients
    grad_alpha = np.mean(p - y)
    grad_beta = np.mean((p - y) * x)

    # Update parameters
    alpha = alpha - learning_rate * grad_alpha
    beta = beta - learning_rate * grad_beta

    # Status update
    if i % (num_epochs // 10) == 0:
        print(f"Epoch: {i}  Estimated alpha: {alpha:.6f}  Estimated beta: {beta:.6f}")

# Output the final estimated parameters
print(f"\nFinal estimated alpha: {alpha:.6f}")
print(f"Final estimated beta:  {beta:.6f}")
```

__Estimate the model with scikit-learn's `LogisticRegression`__

```python
model = LogisticRegression(C=np.inf, max_iter=1000)  # C=np.inf disables regularisation
model.fit(x.reshape(-1, 1), y)

print(f"Intercept: {model.intercept_[0]:.6f}")
print(f"Slope:     {model.coef_[0][0]:.6f}")
```
